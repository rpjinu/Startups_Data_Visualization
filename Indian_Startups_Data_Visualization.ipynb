```python
#extract year from funding date column
df['Funding_Year']=pd.to_datetime(df['Funding_Date']).dt.year
```

#***Data Visualization:-***

```python
# Sector distribution of startup
plt.figure(figsize=(10, 6))
df['Sector'].value_counts().plot(kind='bar', color='skyblue')
plt.title("Startup Distribution by Sector")
plt.xlabel("Sector")
plt.ylabel("Number of Startups")
plt.xticks(rotation=45)
plt.show()
```

```python
# State distribution of startup
plt.figure(figsize=(10, 6))
sns.countplot(x=df['State'], order=df['State'].value_counts().index, palette='viridis')
plt.title("Startup Distribution by State")
plt.ylabel("Number of Startups")
plt.xlabel("State")
plt.show()
```

```python
# Funding stage distribution
plt.figure(figsize=(10, 6))
df['Funding_Stage'].value_counts().plot(kind='bar', color='skyblue')
plt.title("Funding Stage Distribution")
plt.xlabel("Funding Stage")
plt.ylabel("Number of Startups")
plt.show()
```

```python
# ESG Score vs Diversity Index
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='ESG_Score', y='Diversity_Index', hue='Sector', palette='rainbow')
plt.title("ESG Score vs Diversity Index")
plt.xlabel("ESG Score")
plt.ylabel("Diversity Index")
plt.legend(title="Sector", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# Yearly trend of startups founded
plt.figure(figsize=(10, 6))
df['Founded_Year'].value_counts().sort_index().plot(kind='line', marker='o', color='teal')
plt.title("Number of Startups Founded Over the Years")
plt.xlabel("Year")
plt.ylabel("Number of Startups")
plt.grid()
plt.show()
```

```python
#Funding Trends Over the Years
plt.figure(figsize=(12, 6))
funding_trends = df.groupby('Funding_Year')['Amount_Raised'].sum()
funding_trends.plot(kind='line', marker='o', color='blue')
plt.title("Total Funding Amount Raised Over the Years")
plt.xlabel("Year")
plt.ylabel("Total number of Funding company get")
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
df['Funding_Stage'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=plt.cm.Paired.colors, startangle=140)
plt.title("Distribution of Funding Stages")
plt.ylabel("")  # Hides the y-label
plt.show()
```

```python
# Average Funding Amount by State
state_funding = df.groupby('State')['Amount_Raised'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(10, 6))
state_funding.plot(kind='bar')
plt.title("Top 10 States by Average Funding Amount")
plt.xlabel("State")
plt.ylabel("Average Funding Amount")
plt.xticks(rotation=45)
plt.show()
```

```python
#growth rate across sector
plt.figure(figsize=(12, 6))
sector_growth = df.groupby('Sector')['Growth_Rate'].mean().sort_values(ascending=False).head(10)
sector_growth.plot(kind='barh')
plt.title("Top 10 Sectors by Average Growth Rate")
plt.xlabel("Average Growth Rate (%)")
plt.ylabel("Sector")
plt.show()
```

```python
#Number of Patents Across Sectors
plt.figure(figsize=(10, 6))
sector_patents = df.groupby('Sector')['Patents'].sum().sort_values(ascending=False).head(10)
sector_patents.plot(kind='bar', color='coral')
plt.title("Top 10 Sectors by Total Patents")
plt.xlabel("Sector")
plt.ylabel("Number of Patents")
plt.xticks(rotation=45)
plt.show()
```

```python
#Average ESG Score by Sector
plt.figure(figsize=(10, 6))
esg_score = df.groupby('Sector')['ESG_Score'].mean().sort_values(ascending=False).head(10)
esg_score.plot(kind='bar', color='skyblue')
plt.title("Top 10 Sectors by Average ESG Score")
plt.xlabel("Sector")
plt.ylabel("Average ESG Score")
plt.xticks(rotation=45)
plt.show()
```

```python
#Exit Status by Funding Stage
plt.figure(figsize=(10, 6))
sns.countplot(x='Funding_Stage', hue='Exit_Status', data=df, palette='Set2')
plt.title("Exit Status by Funding Stage")
plt.xlabel("Funding Stage")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
#Sector wise growth average
# Grouping by sector to calculate the average growth rate
sector_growth_avg = df.groupby('Sector')['Growth_Rate'].mean().sort_values(ascending=False)

# Plotting the bar chart
plt.figure(figsize=(10, 8))
sector_growth_avg.plot(kind='barh', color='skyblue', edgecolor='black')
plt.title("Average Growth Rate by Sector", fontsize=16)
plt.xlabel("Average Growth Rate (%)", fontsize=14)
plt.ylabel("Sector", fontsize=14)
plt.show()
```

```python
# Grouping data by Sector and Exit Status
sector_exit_status = df.groupby(['Sector', 'Exit_Status']).size().unstack()

# Plotting the double bar chart
plt.figure(figsize=(12, 8))
sector_exit_status.plot(kind='bar', color=['lightcoral', 'lightgreen'], edgecolor='black', width=0.8)

# Adding chart details
plt.title("Exit Status Across Sectors", fontsize=16)
plt.xlabel("Sector", fontsize=14)
plt.ylabel("Number of Startups", fontsize=14)
plt.legend(title="Exit Status", labels=['Not Exited', 'Exited'], fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Grouping by Sector and summing up the Customer Base Size
sector_customer_base = df.groupby('Sector')['Customer_Base_Size'].sum().sort_values(ascending=False)

# Finding the sector with the highest customer base
top_sector = sector_customer_base.idxmax()
top_value = sector_customer_base.max()

print(f"The sector with the highest customer base is '{top_sector}' with a total customer base size of {top_value}.")

# Plotting the top sectors by customer base
plt.figure(figsize=(12, 6))
sector_customer_base.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title("Total Customer Base Size Across Sectors", fontsize=16)
plt.xlabel("Sector", fontsize=14)
plt.ylabel("Total Customer Base Size", fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Distribution of startups across cities
city_counts = df['City'].value_counts()

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(x=city_counts.index, y=city_counts.values, palette='viridis')
plt.title('Distribution of Startups Across Cities')
plt.xlabel('City')
plt.ylabel('Number of Startups')
plt.xticks(rotation=90)
plt.show()
```

```python
# Group by City and sum the Amount_Raised
city_funding = df.groupby('City')['Amount_Raised'].sum().reset_index()

# Sort the data by Amount_Raised in descending order
city_funding = city_funding.sort_values(by='Amount_Raised', ascending=False)

# Plot
plt.figure(figsize=(8, 6))
sns.barplot(x='City', y='Amount_Raised', data=city_funding, palette='viridis')
plt.title('Total Amount Raised by City', fontsize=16)
plt.xlabel('City', fontsize=14)
plt.ylabel('Total Amount Raised (in USD)', fontsize=14)
plt.xticks(rotation=90)  # Rotate city names for better readability
plt.tight_layout()  # Adjust layout to prevent overlapping
plt.show()
```

```python
# Group by City and sum the Investors_Count
city_investors = df.groupby('City')['Investors_Count'].sum().reset_index()

# Sort the data by Investors_Count in descending order
city_investors = city_investors.sort_values(by='Investors_Count', ascending=False)

# Plot
plt.figure(figsize=(8, 6))
sns.barplot(x='City', y='Investors_Count', data=city_investors, palette='magma')
plt.title('Total Investors Count by City', fontsize=16)
plt.xlabel('City', fontsize=14)
plt.ylabel('Total Investors Count', fontsize=14)
plt.xticks(rotation=90)  # Rotate city names for better readability
plt.tight_layout()  # Adjust layout to prevent overlapping
plt.show()
```

```python
# Group by City and sum the Valuation_Post_Funding
city_valuation = df.groupby('City')['Valuation_Post_Funding'].sum().reset_index()

# Sort the data by Valuation_Post_Funding in descending order
city_valuation = city_valuation.sort_values(by='Valuation_Post_Funding', ascending=False)

# Plot
plt.figure(figsize=(8, 6))
sns.barplot(x='City', y='Valuation_Post_Funding', data=city_valuation, palette='viridis')
plt.title('Total Valuation by City', fontsize=16)
plt.xlabel('City', fontsize=14)
plt.ylabel('Total Valuation (in USD)', fontsize=14)
plt.xticks(rotation=90)  # Rotate city names for better readability
plt.tight_layout()  # Adjust layout to prevent overlapping
plt.show()
```

```python
# Group by Sector and Investment_Type, and count the occurrences
sector_investment = df.groupby(['Sector', 'Investment_Type']).size().reset_index(name='Count')

# Pivot the data for better visualization
pivot_table = sector_investment.pivot(index='Sector', columns='Investment_Type', values='Count')

# Plot unstacked (grouped) bar chart
plt.figure(figsize=(8, 6))
pivot_table.plot(kind='bar', colormap='viridis', figsize=(12, 8))
plt.title('Sector vs Investment Type (Unstacked Bar Chart)', fontsize=16)
plt.xlabel('Sector', fontsize=14)
plt.ylabel('Count of Investment Types', fontsize=14)
plt.xticks(rotation=45, ha='right')  # Rotate sector names for better readability
plt.legend(title='Investment Type', bbox_to_anchor=(1.05, 1), loc='upper left')  # Move legend outside
plt.tight_layout()  # Adjust layout to prevent overlapping
plt.show()
```

```python
# Group by Sector and sum the Valuation_Post_Funding
sector_valuation = df.groupby('Sector')['Revenue'].sum().reset_index()

# Sort the data by Valuation_Post_Funding in descending order
sector_valuation = sector_valuation.sort_values(by='Revenue', ascending=False)

# Plot
plt.figure(figsize=(8, 6))
sns.barplot(x='Sector', y='Revenue', data=sector_valuation, palette='viridis')
plt.title('Total Revenue by Sector', fontsize=16)
plt.xlabel('Sector', fontsize=14)
plt.ylabel('Total Revenue(in USD)', fontsize=14)
plt.xticks(rotation=45, ha='right')  # Rotate sector names for better readability
plt.tight_layout()  # Adjust layout to prevent overlapping
plt.show()
```

```python
# Group by Sector and sum the Employee_Count
sector_employees = df.groupby('Sector')['Number_of_Employees'].sum().reset_index()

# Sort the data by Employee_Count in descending order
sector_employees = sector_employees.sort_values(by='Number_of_Employees', ascending=False)

# Plot
plt.figure(figsize=(8, 6))
sns.barplot(x='Sector', y='Number_of_Employees', data=sector_employees, palette='viridis')
plt.title('Total Employee Count by Sector', fontsize=16)
plt.xlabel('Sector', fontsize=14)
plt.ylabel('Total Employee Count', fontsize=14)
plt.xticks(rotation=45, ha='right')  # Rotate sector names for better readability
plt.tight_layout()  # Adjust layout to prevent overlapping
plt.show()
```
